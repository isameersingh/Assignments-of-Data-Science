# Normalization 

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaled_crime_data=scaler.fit_transform(crime.iloc[:,1:])
scaled_crime_data
```

# Build Cluster Algorithm

```python
from sklearn.cluster import KMeans
```

```python
cluster_new=KMeans(3,random_state=42)
cluster_new.fit(scaled_crime_data)
```

```python
cluster_new.labels_
```

```python
crime["cluster_new"]=cluster_new.labels_
```

```python
crime
```

```python
grouped=crime.groupby('cluster_new')

for name,group in grouped:
   print(name)
   print(group)
```

# wcss method to find the k 

```python
wcss=[]
for i in range(1,11):
    clstr=KMeans(i,random_state=0)
    clstr.fit(scaled_crime_data)
    wcss.append(clstr.inertia_)
plt.plot(range(1,11),wcss)
plt.title("elbow method")
plt.xlabel("no of cluster")
plt.ylabel("wcss")
plt.show
```
