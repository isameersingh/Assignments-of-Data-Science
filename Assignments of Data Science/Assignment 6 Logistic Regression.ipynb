```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
data=pd.read_csv(r"C:\Users\pc\Desktop\Data Science Assignments\Logestic\bank-full.csv")
data
```

```python
from sklearn import preprocessing
```

```python
label_encode=preprocessing.LabelEncoder()
```

```python
data["job"]=label_encode.fit_transform(data["job"])
data["marital"]=label_encode.fit_transform(data["marital"])
data["education"]=label_encode.fit_transform(data["education"])
data["default"]=label_encode.fit_transform(data["default"])
data["housing"]=label_encode.fit_transform(data["housing"])
data["loan"]=label_encode.fit_transform(data["loan"])
data["contact"]=label_encode.fit_transform(data["contact"])
data["month"]=label_encode.fit_transform(data["month"])
data["poutcome"]=label_encode.fit_transform(data["poutcome"])
data["y"]=label_encode.fit_transform(data["y"])
data
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=data.iloc[:,0:16]
y=data.iloc[:,16]
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.83,random_state=7)
```

# model building

```python
classifier = LogisticRegression()
```

```python
classifier.fit(x_train,y_train)
```

```python
y_prdct=classifier.predict(x_test)
```

```python
df=pd.DataFrame({'y':y_test,'y_prdct':y_prdct})
df
```

# confusion metrics

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix = confusion_matrix(y_test,y_prdct)
print (confusion_matrix)
```

```python
from sklearn.metrics import accuracy_score
print ("Accuracy : ", accuracy_score(y_test, y_prdct))
l=classifier.predict_proba(x)
print(l)
```
