# Preparing Model using statsmodel formula

```python
import statsmodels.formula.api as smf
```

```python
model=smf.ols('price~age+km+hp+cc+doors+gears+qtrtax+weight',data=data).fit()
model.summary()
```

# calculating varience inflation factor

```python
rsq_age=smf.ols('age~km+hp+cc+doors+gears+qtrtax+weight',data-data).fit().rsquared
vif_age=1/1-rsq_age
rsq_km=smf.ols('km~age+hp+cc+doors+gears+qtrtax+weight',data=data).fit().rsquared
vif_km=1/1-rsq_km
rsq_hp=smf.ols('hp~km+age+cc+doors+gears+qtrtax+weight',data=data).fit().rsquared
vif_hp=1/1-rsq_hp
rsq_cc=smf.ols('cc~hp+km+age+doors+gears+qtrtax+weight',data=data).fit().rsquared
vif_cc=1/1-rsq_cc
rsq_doors=smf.ols('doors~age+km+cc+gears+qtrtax+weight',data=data).fit().rsquared
vif_doors=1/1-rsq_doors
rsq_gears=smf.ols('gears~age+km+hp+cc+doors+qtrtax+weight',data=data).fit().rsquared
vif_gears=1/1-rsq_gears
rsq_qtrtax=smf.ols('qtrtax~age++km+hp+cc+doors+gears+weight',data=data).fit().rsquared
vif_qtrtax=1/1-rsq_qtrtax
rsq_weight=smf.ols('weight~age+km+hp+cc+doors+gears+qtrtax',data=data).fit().rsquared
vif_weight=1/1-rsq_weight
d1={'variable':['age','km','hp','cc','doors','gears','qtrtax','weight'],'vif':[vif_age,vif_km,vif_hp,vif_cc,vif_doors,vif_gears,vif_qtrtax,vif_weight]}
vif_frame=pd.DataFrame(d1)
vif_frame
```

# test normality of resuduals

```python
model.resid
```

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
qqplot=sm.qqplot(model.resid,line='q')
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

# cook's distance

```python
from statsmodels.graphics.regressionplots import influence_plot
```

```python
model_influence = model.get_influence()
(c, _) = model_influence.cooks_distance
```

```python
#Plot the influencers values using stem plot
fig = plt.subplots(figsize=(20, 7))
plt.stem(np.arange(len(data)), np.round(c, 3))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
#index and value of influencer where c is more than .5
(np.argmax(c),np.max(c))
```
