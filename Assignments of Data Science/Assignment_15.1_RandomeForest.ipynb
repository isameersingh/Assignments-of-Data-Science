```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
data_set=pd.read_csv(r"C:\Users\pc\Desktop\Data Science Assignments\Raandome Forest\Fraud_check.csv")
data_set
```

```python
data_set.info()
```

# Label Encoding

```python
from sklearn import preprocessing
```

```python
lb=preprocessing.LabelEncoder()
```

```python
l=['Undergrad','Marital.Status','Urban']
for i in l:
    data_set[i]=lb.fit_transform(data_set[i])
data_set
```

```python
b=[]
for i in data_set['Taxable.Income']:
    if(i<=30000):
        b.append("Riskey")
    else:
        b.append("Good")
```

```python
data_set['status']=b
data_set.head(50)
```

```python
data_set['status']=lb.fit_transform(data_set['status'])
```

```python
data_set.head(2)
```

# Spliting data using train_test_split

```python
x=data_set.iloc[0:,0:-1]
y=data_set['status']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=101)
```

# model Building

```python
model1=RandomForestClassifier( n_estimators=100,max_depth=5)
```

```python
model1.fit(x_train,y_train)
```

```python
prdct=model1.predict(x_test)
```

```python
prdct
```

```python
df=pd.DataFrame({'y_test':y_test,'prdct':prdct})
df
```

```python
from sklearn.metrics import confusion_matrix,classification_report,plot_confusion_matrix
confusion_matrix(y_test,prdct)
```

```python
print(classification_report(y_test,prdct))
```
