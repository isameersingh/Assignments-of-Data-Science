```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
data_set=pd.read_csv(r"C:\Users\pc\Desktop\Data Science Assignments\Raandome Forest\Company_Data.csv")
data_set.head(10)
```

```python
l=[]
for i in data_set['Sales']:
    if(i<=7.496325):
        l.append('Low Sales')
    else:
        l.append('High Sales')
data_set['Sales_cat']=l
data_set.head()
```

```python
del data_set['Sales']
```

```python
data_set
```

```python
from sklearn import preprocessing
```

```python
lb=preprocessing.LabelEncoder()
```

```python
lst=['ShelveLoc','Urban','US','Sales_cat']
for i in lst:
    data_set[i]=lb.fit_transform( data_set[i])
data_set.head(5)
```

# splitting the Data using train_test_split method

```python
x=data_set.iloc[0:,0:-1]
y=data_set['Sales_cat']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=101)
```

# Model Building

```python
model1=RandomForestClassifier(n_estimators=50, max_depth=5)
model1.fit(x_train,y_train)
```

```python
prdct=model1.predict(x_test)
prdct
```

```python
Df=pd.DataFrame({'y_test':y_test,'prdct':prdct})
Df
```

```python
from sklearn.metrics import confusion_matrix,classification_report,plot_confusion_matrix
confusion_matrix(y_test,prdct)
```

```python
print(classification_report(y_test,prdct))
```
